```python
import nibabel as nib
import pandas as pd
import numpy as np
import os
import sys
import matplotlib.pyplot as plt
import glob
import seaborn as sns
import h5py

import matplotlib.ticker as ticker
from skimage.transform import resize
from sklearn.preprocessing import normalize, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from keras.utils import to_categorical

import cv2 as cv2
from keras.utils.io_utils import HDF5Matrix
from keras.models import Sequential
from keras import optimizers
from keras.layers import Dense, Conv2D, Flatten, MaxPooling2D, Dropout
from keras.callbacks import EarlyStopping
from keras.models import load_model

DIR = 'Brain_Hack/'
WIDTH = 224
HEIGHT = 224
```

### Read h5py data

```python
with h5py.File(DIR + 'BrainHack_quality_train.hdf5','r') as hf:
    print(hf.keys())
with h5py.File(DIR + 'BrainHack_quality_test.hdf5','r') as hf:
    print(hf.keys())
with h5py.File(DIR + 'BrainHack_quality_val.hdf5','r') as hf:
    print(hf.keys())
```

```python
def preprocess(img):
    img = cv2.normalize(img, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
    if len(img.shape)==3:
        img = cv2.resize(img[:,:,0], (224, 224), interpolation = cv2.INTER_AREA)
    if len(img.shape)==2:
        img = cv2.resize(img, (224, 224), interpolation = cv2.INTER_AREA)
    return img

X_train = HDF5Matrix(DIR + 'BrainHack_quality_train.hdf5', 'img', normalizer=preprocess)
y_train_d = HDF5Matrix(DIR + 'BrainHack_quality_train.hdf5', 'labels')
X_valid = HDF5Matrix(DIR + 'BrainHack_quality_val.hdf5', 'img', normalizer=preprocess)
y_valid_d = HDF5Matrix(DIR + 'BrainHack_quality_val.hdf5', 'labels')
X_test = HDF5Matrix(DIR + 'BrainHack_quality_test.hdf5', 'img', normalizer=preprocess)
y_test_d = HDF5Matrix(DIR + 'BrainHack_quality_test.hdf5', 'labels')
```

```python
plt.imshow(X_train[300], cmap='gray')
print(np.max(X_train[300]))
```

#### Selecting labels

```python
y_train = np.array(y_train_d)[:,:2]
y_valid = np.array(y_valid_d)[:,:2]
y_test = np.array(y_test_d)[:,:2]
```

### Assemble model

```python
model = Sequential()
model.add(Conv2D(64, kernel_size=7, activation="relu", input_shape=(224,224,1)))
model.add(MaxPooling2D(pool_size=(7, 7), strides=None, padding='valid'))
model.add(Conv2D(32, kernel_size=5, activation="relu"))
model.add(MaxPooling2D(pool_size=(3, 3), strides=None, padding='valid'))

model.add(Flatten())
model.add(Dropout(0.5))
model.add(Dense(64, activation="relu"))
model.add(Dense(2, activation="softmax"))

model.summary()

adam = optimizers.adam(learning_rate=0.001)
model.compile(optimizer=adam, loss='categorical_crossentropy', metrics=['accuracy'])
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=5)
```

```python
model.fit(X_train, y_train, validation_data=(X_valid, y_valid), epochs=25, shuffle='batch', callbacks=[es])
```

```python
scores = model.evaluate(X_test, y_test, verbose=0)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))

# save model and architecture to single file
model.save('/media/thais-caldeira/data/BrainHack/model1_MRxCT.h5')
```

```python
model = load_model('/media/thais-caldeira/data/BrainHack/model1_MRxCT.h5')
y_pred = model.predict(X_test)
```
